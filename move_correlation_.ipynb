```python
# Import Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
# Read the data
df = pd.read_csv('/content/movies.csv')
```

```python
# Show the data
df.head()
```

```python
# Data info

df.info()
```

```python
# Data describe

df.describe()
```

```python
# Let's see if there missing value

for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for out columns

df.dtypes
```

```python
# Change data type of column

list_int = ['budget','gross','votes']

for col in list_int:
  df[col] = df[col].fillna(0).astype('int64')

df.head()
```

```python
# Extract released_country
df['released_country'] = df['released'].str.extract(r'\((.*?)\)')
df.head()
```

```python
#Change type of released
df['released'] = pd.to_datetime(
    df['released'].str.extract('(.*?)\s*\(', expand=False),
    errors='coerce'
    )

df['released'] = df['released'].dt.strftime('%d-%m-%Y')
df.head(10)
```

```python
# Extract correct year

df['Correct_year'] = df['released'].str.extract(r'(\d{4})')
df.head(10)
```

```python
# Print all dataframe

pd.set_option('display.max_rows', None)
```

```python
# sort value of gross

df = df.sort_values('gross',inplace=False , ascending=False)
```

```python
# Remove duplactie

df.drop_duplicates().head()
```

# ***EDA***

```python
# Scatter plot with budget vs gross
plt.figure(figsize=(12,8))
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross')

plt.xlabel('Gross')

plt.ylabel("Budget")
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="budget", y="gross",data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Let's strat to show correlation
numerical_features = df.select_dtypes(include=np.number)
numerical_features.corr(method='pearson')
```

```python
# Let's strat to show heatmap correlation

correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(15,10))

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Columns")

plt.show()
```

```python
# Change object data to numerical

df_numerized = df

for col_num in df_numerized.columns:
  if(df_numerized[col_num].dtype == 'object'):
    df_numerized[col_num] = df_numerized[col_num].astype('category')
    df_numerized[col_num] = df_numerized[col_num].cat.codes

df_numerized.head()
```

```python
# Compare the data
df.head()
```

```python
# Let's strat to show heatmap correlation

correlation_matrix = df_numerized.corr(numeric_only=True)

plt.figure(figsize=(15,10))

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Columns")

plt.show()
```

```python
df_numerized.corr()
```

```python
# Corr pair
correlation_mat = df_numerized.corr(numeric_only=True)

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort value

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr_variable = high_corr[~(high_corr == 1)]

high_corr_variable
```
